```python
glaxo_df=pd.read_csv("GLAXO.csv")
glaxo_df
```

```python
beml_df=beml_df[['Date','Close']]
beml_df
```

```python
glaxo_df=glaxo_df[['Date','Close']]
glaxo_df
```

```python
##making date column into datetime datatype
glaxo_df['Date']=pd.to_datetime(glaxo_df['Date'])
beml_df['Date']=pd.to_datetime(beml_df['Date'])
```

```python
glaxo_df.dtypes
```

```python
beml_df.dtypes
```

```python
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date'])).drop('Date',axis=1)
glaxo_df                                                                               ##making date column as index
```

```python
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date'])).drop('Date',axis=1)
beml_df                                                                           
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
## matplotlib inline
plt.plot(glaxo_df.Close);
plt.xlabel('Time');
plt.ylabel('Close');
```

```python
plt.plot(beml_df.Close)
plt.xlabel('Time');
plt.ylabel('Close');
```

```python
glaxo_df['gain']=glaxo_df.Close.pct_change(periods=1)
beml_df['gain']=beml_df.Close.pct_change(periods=1)
```

```python
glaxo_df                                                    ##  Calculating gain
```

```python
beml_df
```

```python
## drop first row since it is NaN
glaxo_df=glaxo_df.dropna()
glaxo_df
```

```python
beml_df=beml_df.dropna()
beml_df
```

```python
## plotting the gains
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,4))
plt.plot(glaxo_df.gain);
plt.xlabel('Time');
plt.ylabel('gain');
```

```python
plt.figure(figsize=(7,4))
plt.plot(beml_df.gain);
plt.xlabel('Time');
plt.ylabel('gain');
```

```python
import seaborn as sn
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.suptitle('Distribution',fontsize=15)

sn.distplot(glaxo_df.gain,label='Glaxo');

plt.xlabel('gain');
plt.ylabel('density');
plt.legend();

plt.subplot(1,2,2)
sn.distplot(beml_df.gain,label='Beml');

plt.xlabel('gain');
plt.ylabel('density');
plt.legend()

```

```python
print('Mean:',round(glaxo_df.gain.mean(),4))
print('Standard Deviation:',round(glaxo_df.gain.std(),4))
```

```python
print('Mean:',round(beml_df.gain.mean(),4))
print('Standard Deviation:',round(beml_df.gain.std(),4))
```
